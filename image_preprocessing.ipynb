```python
from torchvision import transforms
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
application_image_directory='/Users/shreyaspatel/Desktop/Concordia University/SEM02/COMP 6721 Applied AI/Project/Part 1/Final_Data/Application Mask Dataset'
images_path_list = os.listdir(application_image_directory)
plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,5)
i=0
for imageFile in images_path_list:
    #print(imageFile)
    if ".DS_Store" not in imageFile:
        #image = Image.open(application_image_directory+"/"+imageFile)
        application_image = plt.imread(application_image_directory+"/"+imageFile)
        application_image = ax[i].imshow(application_image, extent=[0, 300, 0, 300])
        ax[i].axes.yaxis.set_visible(False)
        ax[i].axes.xaxis.set_visible(False)
        i = i+1
plt.show()
```

```python
import PIL.Image
from torchvision import transforms
```

# brightness

```python
plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,5)
i=0
for imageFile in images_path_list:
    #print(imageFile)
    if ".DS_Store" not in imageFile:
        #image = Image.open(application_image_directory+"/"+imageFile)
        application_image = PIL.Image.open(application_image_directory+"/"+imageFile)
        loader_transform_brightness = transforms.ColorJitter(brightness=2)
        application_image_brightness = loader_transform_brightness(application_image)
        application_image_brightness = ax[i].imshow(application_image_brightness, extent=[0, 300, 0, 300])
        ax[i].axes.yaxis.set_visible(False)
        ax[i].axes.xaxis.set_visible(False)
        i = i+1
plt.show()
```

# hue

```python
plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,5)
i=0
for imageFile in images_path_list:
    #print(imageFile)
    if ".DS_Store" not in imageFile:
        #image = Image.open(application_image_directory+"/"+imageFile)
        application_image = PIL.Image.open(application_image_directory+"/"+imageFile)
        loader_transform_hue = transforms.ColorJitter(hue=0.15)
        application_image_hue = loader_transform_hue(application_image)
        application_image_hue = ax[i].imshow(application_image_hue, extent=[0, 300, 0, 300])
        ax[i].axes.yaxis.set_visible(False)
        ax[i].axes.xaxis.set_visible(False)
        i = i+1
plt.show()
```

# flip 

```python
plt.rcParams["figure.figsize"] = [15.00, 15.00]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,5)
i=0
for imageFile in images_path_list:
    #print(imageFile)
    if ".DS_Store" not in imageFile:
        #image = Image.open(application_image_directory+"/"+imageFile)
        application_image = PIL.Image.open(application_image_directory+"/"+imageFile)
        loader_transform_flip = transforms.RandomHorizontalFlip(p=1)
        application_image_flip = loader_transform_flip(application_image)
        application_image_flip = ax[i].imshow(application_image_flip, extent=[0, 300, 0, 300])
        ax[i].axes.yaxis.set_visible(False)
        ax[i].axes.xaxis.set_visible(False)
        i = i+1
plt.show()
```
